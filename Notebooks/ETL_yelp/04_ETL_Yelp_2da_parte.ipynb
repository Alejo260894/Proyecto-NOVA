## reviews Yelp

```python
table_reviews = pd.read_parquet(os.path.join(Ruta_data_procesada_full, "Dataset_Review.parquet"))
table_reviews.head(3)
```

```python
# Separar los elementos de la columna "date" en cuatro columnas
date_split = table_reviews['date'].str.split(' ', expand=True)

# Asignar nombres a las nuevas columnas
date_split.columns = ['fecha', 'hora']

# Dividir la columna "fecha" en columnas separadas para año, mes y día
date_split[['anho', 'mes', 'dia']] = date_split['fecha'].str.split('-', expand=True)

# Unir las nuevas columnas con el DataFrame original
table_reviews = table_reviews.join(date_split[['anho', 'mes', 'dia', 'hora']])

# Eliminar la columna "date"
table_reviews.drop(columns=['date'], inplace=True)
```

```python
table_reviews.head(2)
```

```python
# Identificar filas duplicadas basadas en múltiples columnas
#duplicates = table_reviews.duplicated(subset=['user_id', 'text', 'id_user', 'id_business', 'anho', 'mes', 'dia'], keep=False)
duplicates = table_reviews.duplicated(subset=['user_id', 'id_business', 'anho', 'mes', 'dia'], keep=False)
# Filtrar el DataFrame original para mostrar solo las filas duplicadas
duplicated_rows = table_reviews[duplicates]
duplicated_rows.head(2)
```

```python
table_reviews = table_reviews.drop_duplicates(subset=['user_id', 'id_business', 'anho', 'mes', 'dia'])
table_reviews.info()
```

```python
table_reviews.isnull().sum()
```

```python
# Cargamos los dataframes creados en la etapa anterior de ETL
Dataset_Business = pd.read_parquet(os.path.join(Ruta_data_procesada, "Dataset_Business.parquet"))
UsuarioYelp = pd.read_csv(os.path.join(Ruta_data_preprocesada, "UsuarioYelp.csv"))
```
