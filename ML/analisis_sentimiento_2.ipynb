```python
import pyarrow.csv as pv
import pyarrow.parquet as pq
import pandas as pd
import nltk
from textblob import TextBlob
```

```python
df_business = pd.read_parquet("../Data/data_procesada/restaurantes_unificados.parquet")
df_business
```

```python
df_reviews = pd.read_parquet("../Data/data_procesada/reviews_unificadas.parquet")
df_reviews
```

```python
df_reviews.info()
```

```python
df_business = df_business.drop(columns=['postal_code', 'latitude', 'longitude', 'price_numeric'])
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
def analizar_sentimiento(texto):
    blob = TextBlob(texto)
    sentimiento = blob.sentiment.polarity
    return sentimiento
```

```python
df_reviews['sentimiento'] = df_reviews['text'].apply(analizar_sentimiento)
```

```python
df_reviews
```

```python
# Selecciona las columnas 'restaurant_id' y 'rating'
df_selected = df_reviews[['restaurant_id', 'rating', 'sentimiento']]

# Agrupa por 'restaurant_id' y 'rating', luego calcula la media de 'sentimiento'
analisis_por_estrellas = df_selected.groupby(['restaurant_id', 'rating'])['sentimiento'].mean().reset_index()
```

```python
analisis_por_estrellas
```

```python
def escalar_sentimiento(sentimiento):
    # Escalar el sentimiento de -1 a 1 a un rango de 1 a 10
    sentimiento_escalado = (sentimiento + 1) * 5
    return sentimiento_escalado
```

```python
df_reviews['sentimiento_escalado'] = df_reviews['sentimiento'].apply(escalar_sentimiento)
```

```python
df_reviews = df_reviews.groupby(['restaurant_id', 'rating'])['sentimiento_escalado'].mean().reset_index()
```

```python
df_reviews
```

```python
# Calcular los valores medios de sentimiento y sentimiento_escalado para cada id_business
df_sentimiento_agrupado = df_reviews.groupby('restaurant_id').agg({'sentimiento': 'mean', 'sentimiento_escalado': 'mean'}).reset_index()

```

```python
df_sentimiento_agrupado
```

```python
# Fusionar los valores medios con el DataFrame df_business
df_business_agrupado = df_reviews.merge(df_sentimiento_agrupado, on='restaurant_id', how='left')
```

```python
# Eliminar duplicados despuÃ©s de la fusiÃ³n
df_business_agrupado.drop_duplicates(subset=['restaurant_id'], inplace=True)
```

```python
df_business_agrupado
```

```python
df_ml = df_business.merge(df_sentimiento_agrupado[['restaurant_id', 'sentimiento', 'sentimiento_escalado']], on='restaurant_id', how='left')
```
